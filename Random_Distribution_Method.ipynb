```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_mixing(N, M, num_simulations):
    """
        N : Number of molecules of type A
        M : Number of molecules of type B
        num_simulations : Number of simulations

    Returns:
        ndarray: Array containing the number of molecules of type A on one side of the container after each mixing.
    """
    total_molecules = N + M

    # Randomly distribute molecules after removing the partition
    count_A_left = np.random.randint(0, N + 1, size=num_simulations)

    return count_A_left


N = 10
M = 10
num_simulations = 1000

results = simulate_mixing(N, M, num_simulations)

# Plot a histogram
plt.hist(results, bins=range(N+2), align='left', alpha=0.75)
plt.xlabel('Number of Type A Molecules on One Side')
plt.ylabel(' Number of Occurrences')
plt.title('Distribution of Type A Molecules after Mixing')
plt.xticks(range(N+1))
plt.grid(axis='y', alpha=0.5)
plt.show()

import numpy as np

# The mean and standard deviation
mean = np.mean(results)
std_dev = np.std(results)

print("Mean:", mean)
print("Standard Deviation:", std_dev)

```

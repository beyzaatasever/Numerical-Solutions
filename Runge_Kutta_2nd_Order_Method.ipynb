```python
import matplotlib.pyplot as plt

def f(x, u):
    return x**2 * 2.71828**(-u - 1)

h = 0.01
x = 0
u = 1
results = []

x_max = 3

num_steps = int((x_max - x) / h) + 1

# loop
for i in range(num_steps):

    results.append((x, u))

    # k1
    k1 = f(x, u)

    # k2
    x_temp = x + h
    u_temp = u + h * k1
    k2 = f(x_temp, u_temp)

    # RK2
    u = u + (h / 2) * (k1 + k2)

    # increasing
    x = x_temp

x_v = [point[0] for point in results]
u_v = [point[1] for point in results]

# Plotting
plt.plot(x_v, u_v, label='RK2 Solution', color='green', linestyle='-', linewidth=4)
plt.xlabel("x")
plt.ylabel("u(x)")
plt.title("RK2 Solution")
plt.grid(True, linestyle='--', linewidth=0.5)  # Customize grid
plt.legend(loc='upper left', fontsize=10, frameon=True, framealpha=1, edgecolor='black')  # Legend box
plt.show()
```

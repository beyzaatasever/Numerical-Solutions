```python
#Beyza Atasever#HW12 2355097

# Unit cube parameters with zero boundary conditions and source term f = -10
L = 1.0
N = 6
dx = L / N
dy = dx
dz = dx
f = -10

# I initialized the temperature grid
T = [[[0.0 for _ in range(N+1)] for j in range(N+1)] for i in range(N+1)]

tolerance = 1e-6
max_steps = 10000  # Maximum number of steps

# Function to update the temperature grid
def update_temperature(T, f, dx, dy, dz, N):
    T_new = [[[T[i][j][k] for k in range(N+1)] for j in range(N+1)] for i in range(N+1)]
    for i in range(1, N):
        for j in range(1, N):
            for k in range(1, N):
                T_new[i][j][k] = (
                    (T[i+1][j][k] + T[i-1][j][k]) / dx**2 +
                    (T[i][j+1][k] + T[i][j-1][k]) / dy**2 +
                    (T[i][j][k+1] + T[i][j][k-1]) / dz**2 -
                    f
                ) / (2/dx**2 + 2/dy**2 + 2/dz**2)
    return T_new

# I calculated the difference between two grids
def calculate_difference(T_new, T, N):
    diff = 0.0
    for i in range(N+1):
        for j in range(N+1):
            for k in range(N+1):
                diff += (T_new[i][j][k] - T[i][j][k])**2
    return diff**0.5

# Gauss Seidel
for step in range(max_steps):
    # New temperature grid
    T_new = update_temperature(T, f, dx, dy, dz, N)
    # I calculat0de the difference between new and old temperature grids
    diff = calculate_difference(T_new, T, N)
    # the old grid with the new values
    T = T_new
    # Check for convergence
    if diff < tolerance:
        print(f"Converged after {step+1} steps")
        break
else:
    print("Did not converge")

# Results for temperature distribution
for i in range(N+1):
    for j in range(N+1):
        for k in range(N+1):
            print(f"T[{i}][{j}][{k}] = {T[i][j][k]}")
```
